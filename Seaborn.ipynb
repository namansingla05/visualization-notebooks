```python
#Bar Plot
import seaborn as sns
import matplotlib.pyplot as plt

data = sns.load_dataset("tips") #their are some predefined datatset present in sns in git
print(data)

sns.barplot(data=data,x="day",y="tip",estimator="median") #its default is mean, we cant use count
plt.show()
```

```python
sns.barplot(data=data,x="day",y="tip",hue="sex") #its default is mean, we cant use count
plt.show()
#we can change the order also by usung order attribute order =["sun","sat","fri","thur"]
#the black line is for error we can remove that by using attributwe errorbar=("ci",0) ci is confif=dence interval
```

```python
#scatter plot when number to number data
sns.scatterplot(data=data,x="total_bill",y="tip")
plt.show()

sns.scatterplot(data=data,x="total_bill",y="tip",hue="smoker")
plt.show()

sns.scatterplot(data=data,x="total_bill",y="tip",hue="smoker",size="size",palette="viridis")
plt.legend(bbox_to_anchor=(0.3,-1,1,2)) #(x,y,width,height)
plt.show()
```

```python
#heatmap

import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
data = sns.load_dataset("tips")
gp=data.groupby("day",observed=True).agg({"tip":"mean"})
sns.heatmap(gp)
plt.show()
```

```python
#Relational plot its default type are scatater and line
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
data = sns.load_dataset("tips")
sns.relplot(data=data,x="tip",y="total_bill",hue="sex")
plt.show()
sns.relplot(data=data,x="tip",y="total_bill",hue="sex",kind="line")
plt.show()
```

```python

sns.relplot(data=data,x="tip",y="total_bill",hue="sex",col="day")
plt.show()
sns.relplot(data=data,x="tip",y="total_bill",hue="sex",col="day",size="smoker")
plt.show()
```

```python
#swarm plot = thet is similr to scatter and strip scatter work for numerical numerical but strip can work for like day and numerical
#but strip dots overlap but in swarm no overlap is there
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
data = sns.load_dataset("tips")
sns.stripplot(data=data,x="day",y="total_bill",jitter=0.4) #after jitter some still overlapping 
plt.show()
sns.swarmplot(data=data,x="day",y="total_bill")
plt.show()
sns.swarmplot(data=data,x="day",y="total_bill",hue="sex")
plt.show()
sns.swarmplot(data=data,x="day",y="total_bill",hue="sex",dodge=True)
plt.show()
```

```python
#KDE PLOT KDE stands for kernal density estimation(similarto hist wave form of hist)
#It represents the probability density function (PDF) of the data
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
data = sns.load_dataset("tips")
sns.histplot(data=data,x="total_bill")
plt.show()
sns.kdeplot(data=data,x="total_bill")
plt.show()
sns.kdeplot(data=data,x="total_bill",hue="day")
plt.show()
sns.kdeplot(data=data,x="total_bill",hue="day",multiple="stack") #its depend on area
plt.show()
```
