```python
#to plot data from the dataset
import pandas as pd
data=pd.read_csv("employees.csv")
df=pd.DataFrame(data)
print(df)
```

```python

df["Gender"] = df["Gender"].astype(str) #this will not work with nan values
plt.bar(df["Gender"],df["Salary"]) #this does not show the graph properly for that we need to use group by(this will show the max values)
plt.show()
```

```python
g=df.groupby("Gender")["Salary"].sum()

print(g) #we are ingloring nan because we do this after clening the data
#result = df.groupby("Gender")["Salary"].agg(["sum", "mean", "count"]) #You can apply multiple aggregation functions at once using .agg():
#print(result)

plt.bar(g.index,g.values)
plt.show()
```

```python
#scatter plot
import matplotlib.pyplot as plt
import numpy as np

x=np.random.randint(1,10,50)
y=np.random.randint(10,100,50)
colors=np.random.randint(10,100,50)
size=np.random.randint(10,100,50)
plt.scatter(x,y,marker="*",cmap="viridis",c=colors,s=size) #we can use diff cmap values 
plt.colorbar() #which color we are getting
plt.show()
```

```python
#pie Chart(circular chart)
import matplotlib.pyplot as plt
brands = ["Oneplus","Apple", "Samsung","Nokia","Redmi"]
x = [22,35,28,5,10] #popularity
c = ["green","red","blue","pink","orange"]
ex= [0,0,0,0,0.1]
plt.pie(x, labels = brands, colors =c, explode=ex,shadow=True,autopct="%.2f") #explode is for highlighting we can use other attributes(shift tab)
#autopct is for showing the vslue and .2fis for till 2 decimal places and % this is neccessary because this show in percentage
plt.show()

#we can use the data files and we can use the group by and all
```

```python
#Box Plot
import matplotlib.pyplot as plt

l = [25,46,57,25,45,13,29,39,46,99,22,12,14,16,55,6,46,52,55]
#we had studied that
#Q1 = 25%(25 PERCENTILE) ==> (25/100) * (N+1) here N is number of values similarly q3 and orange line is median 
#ans after finding (N+1)/4 we need to find the that number position from sorted order lets say n+1/4 came 2.5 then take the mean of 2nd and 3rd value of data
#LF=Q1 - 1.5(IQR) lower fence(min value) we take min value of the data that is close to lf
#UF=Q3 + 1.5(IQR) Upper fence(max value) we take max value of the data that is close to uf
#IQR=Q3-Q1
plt.boxplot(l)

plt.show()
```
